```python
def primes(num):
    primes = list(range(2, num+1))
    i = 0
    while i < len(primes):
        primes = primes[:i+1] + list(filter(lambda x : x%primes[i] != 0, primes[i+1:]))
        i += 1
    return primes
```

```python
def binary_search(arr,target):
    start = 0 
    end = len(arr)-1
    while start <= end:
        mid  = (start+end)//2
        if arr[mid] ==  target:
            return mid
        elif arr[mid] < target:
            start = mid+1
        elif arr[mid] > target:
            end = mid-1
        
    return -1

lit = [1,2,3,4,5,6,17,18,19,20]
print(binary_search(lit, 7))
```

```python
def bubble_sort(arr:list) :
    n = len(arr)
    for i in range(n):

        for j in range(n-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]   # Swapping
        
    return arr

bubble_sort([1, 112, 13, 8, 5, 6, 7, 8, 9, 10])
```

```python
import time
from datetime import datetime
import time
import numpy as np

t = 0.08
def task_to_run():
    for i in range(100):
        name = np.random.choice(100000)
        email = str(name) + '@gmail.com'
        passwrd = str(np.random.choice(100000000))
        
        driver.find_element('xpath','/html/body/nav/div/div/ul/li[1]/a').click()
        time.sleep(t)
        driver.find_element('xpath','//*[@id="name"]').send_keys(str(name))
        time.sleep(t)
        driver.find_element('xpath','/html/body/div/form/div[2]/input').send_keys(email)
        time.sleep(t)
        driver.find_element('xpath','//*[@id="password"]').send_keys(passwrd)
        time.sleep(t)
        driver.find_element('xpath','/html/body/div/form/div[4]/button').click()
        time.sleep(t)

# Set the specific time (24-hour format)
specific_time = "12:30:00"  # Example: 2:30 PM

# Convert the time to a `datetime` object
target_time = datetime.strptime(specific_time, "%H:%M:%S").time()

print(f"Waiting to run the task at {specific_time}...")

while True:
    # Get the current time
    now = datetime.now().time()

    # Check if the current time matches the target time
    if now >= target_time:
        task_to_run()
        break

    # Wait a bit before checking again
    time.sleep(1)

```
