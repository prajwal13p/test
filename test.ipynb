```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

df = pd.read_csv('test1.csv',header=1)[['Banknifty.1', 'Nifty.1','finnifty.1', 'midcpnifty.1','sensex.1']].rename(columns={'Banknifty.1':'Banknifty','Nifty.1':'Nifty','finnifty.1':'finnifty','midcpnifty.1':'midcpnifty','sensex.1':'sensex'})
```

```python
driver = webdriver.Edge()
```

```python
driver.get('https://web.sensibull.com/chart?tradingSymbol=NIFTY')
```

```python
driver.switch_to.frame(driver.find_element('xpath','/html/body/div[1]/div/div/div[2]/div[1]/div/iframe'))
```

```python

def date_selector(date):
    time.sleep(0.5)
    driver.find_element(By.XPATH,'/html/body/div[2]/div[3]/div[1]/div/div[2]/div/button').click()
    time.sleep(0.5)
    for j in range(10):
        driver.find_element('xpath','//*[@id="overlap-manager-root"]/div/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div/div/div/span/span[1]/input').send_keys(Keys.BACK_SPACE)
    time.sleep(0.5)
    driver.find_element('xpath','/html/body/div[5]/div/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div/div/div/span/span[1]/input').send_keys(date)
    time.sleep(0.5)
    driver.find_element('xpath','/html/body/div[5]/div/div/div[1]/div/div[4]/div/span/button').click()

```

```python
for i in range(len(df[df.columns[0]].to_list())):
    ind = df.columns.to_list()[0]
    dat = df[ind][i]
    print(dat,ind)
    date_selector(dat)
    time.sleep(3)
    # i = i+1
```

```python
ind = df.columns.to_list()[1]
dat = df[ind][i]
print(dat,ind)
date_selector(dat)
i = i+1
```
